```python
import numpy as np
import matplotlib
import pandas as pd
import sklearn
from sklearn.linear_model import LinearRegression
import torch

def check_installations():
    # Print Numpy version and create a simple array
    print(f"Numpy version: {np.__version__}")
    numpy_array = np.array([1, 2, 3])
    print(f"Numpy array: {numpy_array}")

    # Print Matplotlib version and plot the numpy array
    print(f"Matplotlib version: {matplotlib.__version__}")
    matplotlib.pyplot.plot(numpy_array)
    matplotlib.pyplot.title('Matplotlib Plot')
    matplotlib.pyplot.show()

    # Print Pandas version and create a simple DataFrame
    print(f"Pandas version: {pd.__version__}")
    pandas_df = pd.DataFrame(data=numpy_array, columns=['Numbers'])
    print(f"Pandas DataFrame:\n{pandas_df}")

    # Print scikit-learn version and create a Linear Regression model
    print(f"Scikit-learn version: {sklearn.__version__}")
    model = LinearRegression()
    print(f"Scikit-learn LinearRegression model: {model}")

    # Print PyTorch version and create a tensor
    print(f"PyTorch version: {torch.__version__}")
    torch_tensor = torch.tensor([1, 2, 3])
    print(f"PyTorch Tensor: {torch_tensor}")

check_installations()
```
